##***EJERCICIO 1***

Simular el lanzamiento de un dado equilibrado de 6 caras.


*   Mostrar la probabilidad aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
# simular lanzamientos de dados
def lanzar_dados(cant_dados, caras_dados):
    resultado = np.random.randint(1, caras_dados + 1, cant_dados)
    return resultado

dados = 10000
caras = 6

# realizar lanzamientos
resultados = lanzar_dados(dados, caras)
# contabilizar tiradas de 2, 3 y 5
count_dos = (resultados == 2).sum()
count_tres = (resultados == 3).sum()
count_cinco = (resultados == 5).sum()
prob_dos = count_dos / dados
prob_tres = count_tres / dados
prob_cinco = count_cinco / dados


print('Frecuencia de 2:', count_dos)
print('Probabilidad de 2:', prob_dos)
print('Frecuencia de 3:', count_tres)
print('Probabilidad de 3:', prob_tres)
print('Frecuencia de 5:', count_cinco)
print('Probabilidad de 5:', prob_cinco)
```

```python
# graficar los resultados
plt.figure(figsize=(10, 6))
plt.hist(resultados,
         bins=np.arange(0.5, caras+1.5),
         align='mid', edgecolor='black')

plt.show()
```

## ***EJERCICIO 2***

Simular el lanzamiento de un dado NO eqilibrado o cargado de 6 caras, de manera que las chances o probabilidades de salir 1 o 2 sean el doble de obtener otro resultado (3,4,5 o 6)


*   Mostrar la probabilidad aproximada de obtener el número 2, 3 y 5
*   Graficar el histograma de las probabilidades



```python
# simular lanzamientos de dados cargados
def lanzar_dados_cargados(cant_dados, caras_dados):
    resultados = []
    for i in range(cant_dados):
        num = np.random.random()
        if num < 0.5:
            resultado = np.random.choice([1, 2])
        else:
            resultado = np.random.choice(range(3, caras_dados + 1))
        resultados.append(resultado)
    resultados = np.array(resultados)
    return resultados

dados = 10000
caras = 6

# realizar lanzamientos
resultados_cargados = lanzar_dados_cargados(dados, caras)

# contabilizar tiradas de 2, 3 y 5
count_dos_cargados = (resultados_cargados == 2).sum()
count_tres_cargados = (resultados_cargados == 3).sum()
count_cinco_cargados = (resultados_cargados == 5).sum()
# calcular la probabilidad de 2, 3 y 5
prob_dos_cargados = count_dos_cargados / dados
prob_tres_cargados = count_tres_cargados / dados
prob_cinco_cargados = count_cinco_cargados / dados

print("Resultados lanzamientos cargados:")
print('Frecuencia de 2:', count_dos_cargados)
print("Probabilidad de obtener 2:", prob_dos_cargados)
print('Frecuencia de 3:', count_tres_cargados)
print("Probabilidad de obtener 3:", prob_tres_cargados)
print('Frecuencia de 5:', count_cinco_cargados)
print("Probabilidad de obtener 5:", prob_cinco_cargados)
```

```python
# graficar los resultados
plt.figure(figsize=(10, 6))
plt.hist(resultados_cargados,
         bins=np.arange(0.5, caras+1.5),
         align='mid', edgecolor='black')

plt.show()
```

## EJERCICIO 3

 Realizar la simulación del juego {piedra, papel, tijera} donde el resultado de esta simulación deberá ser lo que mi oponente elije. No debe considerarse la simulación de mi elección.



*   Calcular la probabilidad aproximada que mi oponente elija, 'piedra', 'papel', o 'tijera', en N jugadas.
*   Realizar un grafico de barra mostrando estas probabilidades. Concluir algo sobre las probabilidades de estos resultados, cuando el número de jugadas se incrementa bastante.




```python
# simular elección 'piedra', 'papel' o 'tijera'
def ppt(n_jugadas):
    resultados = []
    for i in range(n_jugadas):
        resultado = np.random.choice(['piedra', 'papel', 'tijera'])
        resultados.append(resultado)
    resultados = np.array(resultados)
    return resultados

jugadas = ppt(30)

# contabilizar elecciones
count_piedra = (jugadas == 'piedra').sum()
count_papel = (jugadas == 'papel').sum()
count_tijera = (jugadas == 'tijera').sum()
# calcular la probabilidad
prob_piedra = count_piedra / len(jugadas)
prob_papel = count_papel / len(jugadas)
prob_tijera = count_tijera / len(jugadas)

print("En 30 jugadas:")
print("Frecuencia de 'piedra'", count_piedra)
print("Probabilidad de elegir 'piedra':", prob_piedra)
print("Frecuencia de 'papel'", count_papel)
print("Probabilidad de elegir 'papel':", prob_papel)
print("Frecuencia de 'tijera'", count_tijera)
print("Probabilidad de elegir 'tijera':", prob_tijera)
```

```python
# graficar los resultados
plt.figure(figsize=(10, 6))
plt.hist(jugadas, bins=3,
         align='mid', edgecolor='black')

plt.xlabel('Jugada')
plt.ylabel('Frecuencia')
plt.title('Simulación de 30 jugadas')
plt.show()
```

 En el gráfico anterior que se observa que cuando el numero de 'jugadas' es bajo la probabilidad de que alguno de los valores aparezca más veces es mayor. Al aumentar el número de 'jugadas', las probabilidades tinden a igualarse al valor esperado de 0.33 para cada opción.

```python
# simular un número mayor de jugadas
muchas_jugadas = ppt(1000)

# graficar los resultados
plt.figure(figsize=(10, 6))
plt.hist(muchas_jugadas, bins=3,
         align='mid', edgecolor='black')

plt.xlabel('Jugada')
plt.ylabel('Frecuencia')
plt.title('Simulación de 1000 jugadas')
plt.show()
```

## EJERCICIO 6

Se realiza una encuesta a N usuario donde se registran los siguientes datos.

['DNI','GASTOS MENSUALES DE HOGAR','INGRESOS MENSUALES POR HOGAR', 'CANTIDAD DE MIEMBROS QUE VIVEN POR HOGAR','GASTO MENSUAL EN LINEA TELEFÓNICA', 'GASTO MENSUAL POR PAGO DE INTERNET' ]

Realizar la simulación de la carga de datos utilizando la siguiente información adicional:


1.   El 70 % de los encuestados son jóvenes entre 15 y 21 años, el 10% son mayores de 60 años. Sugerencia, utilizar esto para generar los DNI
2.   Los encuestados son de la provincia de Cordoba. Además el 80 % de la población de esa provincia tiene un grupo familiar de 5 personas. Con ingresos mensuales promedio de $258.000
2.   Cerca del 80 % de los encuestados Gastan el 90% de sus ingresos, el resto gasta el 80% de sus ingresos.Ayuda, para simular la cantidad de miembros utilizar la distribución normal.
3.  Del gasto Total por hogar se sabe que se abona un 15 % en telefonía y un 7.8% en servicio de internet.

Con la información proporcionada, realizar la simulación de las encuestas de 100 y 1000 personas.

*   Calcular el valor medio y desvio estandar de los gastos mensuales, ingresos mensuales y pagos de servicios.
*   Plotear los datos de los ingresos , y gastos en gráficos de barra por separado y evidenciar la media y desvio estandar.
*   Calcular la proporciona de jóvenes menores a 21 encuestados. Mostrar sus datos de grupos familiares, gastos en servicios de internet y telefonía. Concluir algo en base a esta información.
*   Realizar un histograma con el dato del grupo familiar, y concluir algo acerca de la distribución de probabilidad.




```python
# Desactivar las advertencias de asignación en cadena
pd.options.mode.chained_assignment = None

# Función para simular N encuestas
def simular_encuesta(N):

    # Generar DNI aleatorios para los encuestados
    lista_dni = []
    for i in range(N):
        dni = random.randint(10_000_000, 99_999_999)
        lista_dni.append(dni)

    # Definir la edad de los encuestados
    edades = []
    # 70% jóvenes entre 15 y 21 años
    for i in range(int(N * 0.7)):
        edad = random.randint(15, 21)
        edades.append(edad)

    # 10% mayores de 60 años (61 hasta 100 años)
    for i in range(int(N * 0.10)):
        edad = random.randint(61, 100)
        edades.append(edad)
    # el resto edades al azar entre 22 y 60 años
    while len(edades) < N:
        edad = random.randint(22, 60)
        edades.append(edad)

    # Crear un DataFrame con los DNI y edades
    data = pd.DataFrame({'DNI': lista_dni, 'Edad': edades})

    # Simular ingresos y cantidad de miembros en el hogar
    # Promedio de $258,000 y desvío estándar de $30,000
    data['Ingresos Mensuales'] = np.random.normal(258000, 30000, N)
    # Promedio de 5 miembros y desvío estándar de 1
    data['Cantidad de Miembros'] = np.round(np.random.normal(5, 1, N))

    # Simular gastos mensuales en telefonía e internet
    # Telefonía representa 15% de los ingresos
    data['Gasto Telefonia'] = data['Ingresos Mensuales'] * 0.15
    # Internet representa 7.8% de los ingresos
    data['Gasto Internet'] = data['Ingresos Mensuales'] * 0.078

    # Simular gastos mensuales del hogar
    # El 80% gasta el 90% de sus ingresos
    data['Gastos Mensuales de Hogar'] = data['Ingresos Mensuales'] * 0.9
    # El 20% gasta el 80% de sus ingresos
    data['Gastos Mensuales de Hogar'].iloc[int(N * 0.8):] = \
        data['Ingresos Mensuales'].iloc[int(N * 0.8):] * 0.8

    return data


# Simulación de encuestas para 100 personas
encuesta_100 = simular_encuesta(100)

# Simulación de encuestas para 1000 personas
encuesta_1000 = simular_encuesta(1000)
```
