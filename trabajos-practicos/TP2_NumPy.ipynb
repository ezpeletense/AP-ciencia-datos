**EJERCICIO 2:**

Crear un programa donde se le pida al usuario que ingrese la cantidad de elementos de una lista de números reales positivos. Luego Convertir esa lista en un vector de Numpy.



```python
print('~~~ Bienvenido al creador de vectores ~~~')

# Pedir al usuario la cantidad de elementos para la lista
cant = int(input('Ingrese la cantidad de elementos: '))

# Comprobar que el número ingresado sea positivo
if cant <= 0:
    print('La cantidad de elementos debe ser mayor a cero.')

else:
    # Inicializar la lista de elementos
    lista_elementos = []

    for i in range(1, cant + 1):
        num = input(f'Ingrese el {i}º elemento: ')
        try:
            numero = float(num)
            if numero <= 0:
                print('Los números deben ser positivos.')
                break
            else:
                lista_elementos.append(numero)
        except ValueError:
            print('Ha ocurrido un error, asegurese de ingresar números.')
            break
    else:
        # Convertir la lista en un vector de NumPy
        vector_lst = np.array(lista_elementos)
        print('Vector de NumPy:', vector_lst)
```

Crear un programa donde el usuario ingrese la cantidad de filas y columnas que tendra una tabla de datos. Luego el programa pedira ingresar los datos de la tabla fila por fila. Todos los datos serán numéricos.

Mostrar la tabla ingresada en formato LISTA de Python, y mostrar la misma tabla en formato array de Numpy.

Solicitar al usuario que ingrese las posiciones de dos filas y realice la suma  de las mismas. Mostrar este vector resultado.

```python
print('~~~ Bienvenido al creador de tablas ~~~')

# Pedir al usuario la cantidad de filas y columnas
filas = int(input('Ingrese la cantidad de filas de la tabla: '))
columnas = int(input('Ingrese la cantidad de columnas de la tabla: '))

# Inicializar la tabla vacía
tabla_datos = []

for i in range(1, filas + 1):
    fila = []
    for j in range(1, columnas + 1):
        dato = float(input(f'Ingrese el número en la fila {i}, columna {j}: '))
        fila.append(dato)
    tabla_datos.append(fila)

tabla_np = np.array(tabla_datos)

print('~~~')
print('Tabla en formato lista:\n', tabla_datos)
print('Tabla en formato array:\n', tabla_np)

print('~~~')
# Pedir al usuario dos filas para sumarlas
fila1 = int(input('Ingrese un número de fila: '))
fila2 = int(input('Ingrese otro número de fila: '))

if 1 <= fila1 <= filas and 1 <= fila2 <= filas:
    suma_filas = tabla_np[fila1 - 1] + tabla_np[fila2 - 1]
    print(suma_filas)
else:
    print('Las filas ingresadas no están en la tabla.')
```

A continuación se muestran los valores de  los siguientes productos:

['arroz', 'harina','fideo','yerba','azucar']=[145.6 , 100 , 89.90 , 700 , 95]

Los valores de estos productos son aproximados de hace dos meses, debido a la inflación y alza de los precios, se vieron afectados de la siguiente manera:
*   Producto arroz , harina, azucar duplicaron su precio
*   Productos restantes incrementaron en un 75% su precio

Mostrar los datos en forma de vector y actualizar sus precios, de manera que se pueda comparar ambos vectores.

```python
# Datos iniciales
productos = ['arroz', 'harina', 'fideos', 'yerba', 'azucar']
precios = [145.6, 100, 89.90, 700, 95]

def actualizar_precios(precio, porcentaje):
    return round((precio * porcentaje), 1)

# Aplicar los aumentos de precios correspondientes
precios_act = []

for i in range(len(productos)):
    if productos[i] in ['arroz', 'harina', 'azucar']:
        precios_act.append(actualizar_precios(precios[i], 2))
    else:
        precios_act.append(actualizar_precios(precios[i], 1.75))

productos_np = np.array(productos)
precios_np = np.array(precios)
precios_act_np = np.array(precios_act)

tabla_productos = np.column_stack([productos_np, precios_np, precios_act_np])

print(tabla_productos)
```

Dado una matriz de datos, dividir el 70% de filas en un array_entrenamiento y el otro 30% en otro array_testeo. Esta distribución de filas de la matriz inicial, debe ser aleatoria. Mostrar las matrices al ser modificadas por el comando np.random.shuffle('matriz'). Finalmente mostrar los array_entrenamiento y array_testeo.

```python
dataset = np.array([[25, 1, 7, 100, 1],
                    [30, 2, 5, 120, 0],
                    [22, 1, 6, 80, 1],
                    [28, 1, 6, 90, 0],
                    [35, 2, 4, 130, 1],
                    [32, 2, 6, 110, 1],
                    [26, 1, 8, 95, 1],
                    [24, 1, 5, 85, 0],
                    [29, 2, 7, 115, 1],
                    [31, 2, 6, 105, 0]])

print(dataset)
```

```python
# Mezclar las filas de la matriz para obtener una distribución aleatoria de los datos
np.random.shuffle(dataset)

print('Dataset modificado por np.random.shuffle:\n', dataset)
print(dataset.shape)

# Obtener número de filas de la matriz
num_filas = dataset.shape[0]

# Calcular filas de entrenamiento y testeo
num_filas_en = int(0.7 * num_filas)

# Dividir la matriz en array_entrenamiento y array_testeo
array_entrenamiento = dataset[:num_filas_en, :]
array_testeo = dataset[num_filas_en:, :]

print('Array de entrenamiento:\n', array_entrenamiento)
print()
print('Array de testeo:\n', array_testeo)
```

# ***EJERCICIOS PARTE 3***

Realizar las gráficas de las siguientes expresiones matemáticas. Determinar el dominio del eje x adecuado para plotear las funciones de manera que se visualice su comportamiento.


*   $y= 3x -2$
*   $y= 2x^2+4x+2 $
*   $y= \vert x\vert= \begin{cases}-x & x < 0 \\ x & x\ge 0                       \end{cases}$
*   $y= 1/x$
*   $y=\sqrt x$




```python
# Crear datos para el eje x
x = np.linspace(-10, 10, 400)

# Definir la expresión matemática
y = 3*x - 2

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 3x - 2', color='blue')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = 3x - 2$')

# Línea vertical en x = 0
plt.axvline(0, color='gray', linestyle='-.', linewidth=0.8)
# Línea horizontal en x = 0
plt.axhline(0, color='gray', linestyle='-.', linewidth=0.8)

plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Crear datos para el eje x
x = np.linspace(-20, 20, 400)

# Definir la expresión matemática
y = 2*x**2 + 4*x + 2

# Crear el gráfico
plt.figure(figsize=(8, 6))
plt.plot(x, y, label='y = 2x^2 + 4x + 2', color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = 2x^2 + 4x + 2$')

# Línea vertical en x = 0
plt.axvline(0, color='k', linestyle='--', linewidth=0.8)

plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Definir el dominio (rango de valores de x)
x1 = np.linspace(-5, 0, 400)  # Para x < 0
x2 = np.linspace(0, 5, 400)   # Para x >= 0

# Calcular la función y = |x| para cada intervalo
y1 = -x1
y2 = x2

# Crear el gráfico
plt.plot(x1, y1, label='$y = -x$ para $x < 0$')
plt.plot(x2, y2, label='$y = x$ para $x \geq 0$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = |x|$')

# Línea vertical en x = 0
plt.axvline(0, color='k', linestyle='--', linewidth=0.8)

plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Definir el dominio (rango de valores de x) excluyendo x = 0
x = np.linspace(-40, 40, 400)

# Calcular la función y = 1/x para cada valor en el dominio
y = 1 / x

# Crear el gráfico
plt.plot(x, y, label='$y = \\frac{1}{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = \\frac{1}{x}$')

# Línea vertical en x = 0
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
# Línea horizontal en y = 0
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)

# Establecer los límites de los ejes y para una mejor visualización
plt.ylim(-5, 5)
plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```

```python
# Definir el dominio (rango de valores de x)
x = np.linspace(0, 10, 400)

# Calcular la función y = sqrt(x) para cada valor en el dominio
y = np.sqrt(x)

# Crear el gráfico
plt.plot(x, y, label='$y = \sqrt{x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Gráfico de $y = \sqrt{x}$')

# Línea vertical en x = 0
plt.axvline(0, color='gray', linestyle='--', linewidth=0.8)
# Línea horizontal en y = 0
plt.axhline(0, color='gray', linestyle='--', linewidth=0.8)

plt.legend()
plt.grid(True)

# Mostrar el gráfico
plt.show()
```
